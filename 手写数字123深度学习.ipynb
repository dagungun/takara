```python
import matplotlib.pyplot as plt
import pathlib
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import IPython.display as display
```

# 训练数据集准备

```python
data_root = pathlib.Path("D:/python/num_data/")
print(data_root)
```

```python
all_image_paths = list(item for item in data_root.iterdir())
```

```python
all_image_paths = [str(path) for path in all_image_paths]
print(all_image_paths)
```

```python
all_image = np.empty((0,28,28))
print(type(all_image))
np.shape(all_image)
```

```python
for i in all_image_paths:
    img_plt = plt.imread(i)
    img_plt = np.reshape(img_plt,(1,28,28))
    print(img_plt.shape)
    all_image = np.append(all_image, img_plt,axis = 0)
    print("all_image", all_image.shape)
```

```python
all_image_labels = [1,2,1,2,3,2,3,2,1,2,3,2,1,1,1,3,2,2,3,1,1,2,3,3,2,3,1,1,2,3]
```

```python
all_image_labels = list(item-1 for item in all_image_labels)
```

```python
all_image_labels = np.array(all_image_labels)
print(len(all_image_labels))
print(all_image_labels)
```

```python
train_images = all_image
train_labels = all_image_labels
```

```python
plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
train_images = train_images / 255.0
```

```python
plt.figure(figsize=(5,6))
for i in range(30):
    plt.subplot(5,6,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(train_labels[i])
plt.show()
```

# 开始训练模型

```python
np.random.seed(42)
tf.random.set_seed(42)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(3)
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(train_images, train_labels, epochs=10)
```

```python
def plot_image(i, predictions_array, true_label, img):
  predictions_array, true_label, img = predictions_array, true_label[i], img[i]
  plt.grid(False)
  plt.xticks([])
  plt.yticks([])

  plt.imshow(img, cmap=plt.cm.binary)

  predicted_label = np.argmax(predictions_array)
  if predicted_label == true_label:
    color = 'blue'
  else:
    color = 'red'

  plt.xlabel("{} {:2.0f}% ({})".format(predicted_label+1,
                                100*np.max(predictions_array),
                                true_label+1),
                                color=color)
```

```python
i = 0
plot_image(i, predictions[i], test_labels, test_images)
plt.show()
```

```python
num_rows = 4
num_cols = 5
num_images = num_rows*num_cols
plt.figure(figsize=(2*2*num_cols, 2*num_rows))
for i in range(num_images):
  plt.subplot(num_rows, 2*num_cols, 2*i+1)
  plot_image(i, predictions[i], test_labels, test_images)
plt.tight_layout()
plt.show()
```
